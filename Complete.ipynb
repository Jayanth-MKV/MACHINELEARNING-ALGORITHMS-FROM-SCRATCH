# **LINEAR REGRESSION USING GRADIENT DESCENT**

Linear Regression is a regression technique for the prediction of the continuous target variables

###**Using Gradient Descent**

>$E = (y - ̄y ) $

>$MSE = \frac{1}{m} \sum_{}E^2  $

>$\frac{δMSE}{δa_0} = \frac{-2}{m} ∑E $

>$a_{0} = a_{0} +\alpha . \frac{2}{m}\sum_{i=1}^{m}(y-y')$

>$\frac{δMSE}{δa_1} = \frac{-2}{m} ∑Ex $

>$a_{1} = a_{1} +\alpha . \frac{2}{m}\sum_{i=1}^{m}(y-y').x' $


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,101,size=200)
y=3*np.power(x,1)+5
```

```python
print(x)
print(y)
```

```python
class LinearRegression:
  def __init__(self,lr,e):
    self.a0=np.random.random()
    self.a1=np.random.random()
    self.lr=lr
    self.e=e
    self.itr={}
    self.mx=[]
    self.ms=[]

  def predict(self,x):
    return self.a0+(self.a1)*x

  def mse(self,X,Y,y):
    m=len(X)
    return (1/m)*np.sum(np.power(Y-y,2))

  def fit(self,X,Y):
    l=self.lr
    itr=0
    m=len(X)
    while(itr<self.e):
      y=self.predict(X)
      err=(Y-y)
      self.ms.append(self.mse(X,Y,y))
      self.mx.append(self.a1)
      if(itr%10000==0 and itr<self.e):
        self.itr[itr]=self.mse(X,Y,y)
      d1=(-2/m)*np.sum(err)
      d2=(-2/m)*np.sum(err*x)
      self.a0=(self.a0)-l*d1
      self.a1=(self.a1)-l*d2
      itr=itr+1
    return [self.a0,self.a1]

  def plot(self):
    data=self.itr
    i=list(data.keys())
    error=list(data.values())
    plt.plot(i,error)
    plt.title("Iterations vs Errors")
    plt.xlabel('No of Iterations')
    plt.ylabel('Error')

  def plotD(self,x,y):
    c=[]
    p=np.arange(0,5,0.1)
    for s in p:
      y2=x*s+self.a0
      c.append(self.mse(x,y,y2))

    plt.scatter(p,c)
    plt.plot(self.mx,self.ms,color="red")
    plt.show()
```

```python
model=LinearRegression(0.0001,100000)
[a0,a1] = model.fit(x,y)
print(a0)
print(a1)
```

```python
model.plotD(x,y)
```

```python
model.plot()
```

```python
res=model.predict(x)
plt.plot(x,res,color="blue")
plt.scatter(x,y,color="red")
plt.show()
```
