```python
import pandas as pd
import numpy as np

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

class KNN:
    def _init_(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X):
        y_pred = []
        for x in X:
            distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
            k_indices = np.argsort(distances)[:self.k]
            k_neighbors = [self.y_train[i] for i in k_indices]
            y_pred.append(max(set(k_neighbors), key = k_neighbors.count))
        return np.array(y_pred)

# Load the dataset using pandas
data = pd.read_csv("data.csv")

# Split the data into training and testing sets
train_data = data[:8]
test_data = data[8:]

train_X = train_data.iloc[:, :3].values
train_y = train_data.iloc[:, 3].values
test_X = test_data.iloc[:, :3].values
test_y = test_data.iloc[:, 3].values

# Train the KNN model with k=3
knn = KNN(3)
knn.fit(train_X, train_y)

# Make predictions on the test set
predictions = knn.predict(test_X)

# Calculate the accuracy of the model
accuracy = np.mean(predictions == test_y)
print("Accuracy:", accuracy)
```
