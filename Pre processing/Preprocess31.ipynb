**mean,median,sd using numpy module**

**MEAN** :

The mean of a set of `n` numbers is:

$$\bar{x} = \frac{\sum_{i=1}^n x_i}{n}$$

where `x1`, `x2`, ..., `xn` are the individual values in the set.


```python
me=x.mean()
print(me)
```

**MEDIAN :** 

The median of a set of `n` numbers is:

- If `n` is odd:
    - Sort the numbers in ascending order.
    - The median is the middle number: $x_{\frac{n+1}{2}}$.

- If `n` is even:
    - Sort the numbers in ascending order.
    - The median is the average of the middle two numbers:
 $\frac{x_{\frac{n}{2}}+x_{\frac{n}{2}+1}}{2}$.

```python
md=np.median(x)
print(md)
```

Standard Deviation

```python
sd=np.std(x)
sd
```

**Function to calculate mode**

```python
def mode(x):
  d={}
  k=0
  v=0
  for i in x:
    d[i]=d.get(i,0)+1
    if d[i]>v:
      k=i
      v=d[i]
  # print(d)
  return k

print(mode(x))
```

```python
def mode(x):
  values,cnt=np.unique(x,return_counts=True)
  mo=np.argmax(cnt)
  return values[mo]

print(mode(x))
```

```python
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='black',
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
def f(x, y):
  return np.tan(np.sqrt(x ** 2 + y ** 2))
x=np.random.randint(1,10,[10])
y=np.random.randint(1,10,[10])
x,y = np.meshgrid(x, y)
z = f(x,y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
