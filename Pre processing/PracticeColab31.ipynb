```python
x=np.random.randint(low=60,high=260,size=100)
x
```

```python
me=x.mean()
print(me)
md=np.median(x)
print(md)
mo=stats.mode(x)
print(mo[0])
```

**FINDING MODE**

Method 1

```python
def mode(x):
  values,cnt=np.unique(x,return_counts=True)
  # print(values)
  # print(cnt)
  mo=np.argmax(cnt)
  # print(mo)
  return values[mo]

print(mode(x))
```

Method -2 

```python
def mode(x):
  d={}
  k=0
  v=0
  for i in x:
    d[i]=d.get(i,0)+1
    if d[i]>v:
      k=i
      v=d[i]
  print(d)
  return k

print(mode(x))
```
